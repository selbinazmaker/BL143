```python
# Akıllı Gün Planlayıcı (Kural Tabanlı)

while True:
    try:
        uyku = float(input("Günlük uyku sürenizi (saat) giriniz: "))
        ders = float(input("Günlük ders sürenizi (saat) giriniz: "))
        serbest = float(input("Günlük serbest zamanınızı (saat) giriniz: "))

        # Negatif kontrolü
        if uyku < 0 or ders < 0 or serbest < 0:
            print("Hata: Süreler negatif olamaz. Tekrar deneyiniz.\n")
            continue

        # 24 saat kontrolü
        toplam = uyku + ders + serbest
        if toplam > 24:
            print("Hata: Toplam süre 24 saati geçemez. Tekrar deneyiniz.\n")
            continue

        break  # Tüm kontroller geçildiyse döngüden çık

    except ValueError:
        print("Hata: Lütfen sayısal bir değer giriniz.\n")

# Uyku durumu
if uyku < 6:
    uyku_durumu = "Yetersiz uyku"
elif 6 <= uyku <= 8:
    uyku_durumu = "Normal uyku"
else:
    uyku_durumu = "Fazla uyku"

# Çalışma durumu
if ders >= 6:
    calisma_durumu = "Yoğun çalışma günü"
else:
    calisma_durumu = "Normal çalışma günü"

# Genel değerlendirme
if uyku < 6 and ders >= 6:
    genel_degerlendirme = "Verimsiz gün"
else:
    genel_degerlendirme = "Dengeli gün"

# Çıktılar
print("\n--- Günlük Değerlendirme ---")
print("Uyku durumu:", uyku_durumu)
print("Çalışma durumu:", calisma_durumu)
print("Genel değerlendirme:", genel_degerlendirme)
```
